```python
def extract_verses(chapter):
    output = {}
    result = requests.get("https://www.bible.com/es-ES/bible/128/pro." + str(chapter))
    c = result.content
    soup = BeautifulSoup(c)
    samples = soup.find("div", "chapter")
    for div in samples.find_all("div"):
        for span in div.find_all("span"):
            if 'data-usfm' in span.attrs:
                book, chapter, verse = span.attrs["data-usfm"].split(".")
                key = chapter+"-"+verse
                for subspan in span.find_all("span"):
                    if "content" in subspan.attrs["class"]:
                        if key in output:
                            output[key] = output[key] + " "+ subspan.text
                        else:
                            output[key] = subspan.text
    return output
```

```python
def clean_tokens(tokens, stopwords):
    clean_strings =  [",",".",";", "Los", ":","¿","?", "!","¡","»"," "]
    tokens = [token for token in tokens if token not in clean_strings]
    if stopwords == 1:
        stopwords1 = nltk.corpus.stopwords.words('spanish')
        tokens = [token for token in tokens if token not in stopwords1]
    i = 0
    size_tokens = len(tokens)
    next_tokens = []
    while i < size_tokens:
        if len(tokens[i].split(".")) > 1:
            subtokens = tokens[i].split(".")
            next_tokens =next_tokens+subtokens
        else:
            next_tokens = next_tokens+ [tokens[i]]
        i = i + 1
        
    output = []
    for token in next_tokens:
        for character in clean_strings:
            token = token.replace(character, " ")
        output.append(token)
        
    clean_output = []
    for token in output:
        if token != "" or  token !=  ' ' and len(token)>3:
            clean_output.append(token)
    return clean_output
```

```python
frase_dict = {}
for frase in palabras_frases_comunes:
    if type(frase[0]) is tuple:
        frase_list = list(frase[0])
    else:
        frase_list = [frase[0]]
    frase_texto = " ".join(frase_list)
    frase_dict[frase_texto] = []
    for key in versiculos:
        status = []
        for palabra in frase_list:
            status.append(palabra in versiculos[key])
        if False not in status:
            frase_dict[frase_texto].append(key)
        
```
